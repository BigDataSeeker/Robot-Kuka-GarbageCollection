```python
# Convert annotations to image_size-normalized form
from pathlib import Path
import cv2
from PIL import Image
import os
import numpy as np

class Corrupt_annot(Exception):
    pass


dir = '/content/drive/MyDrive/GarbageClassifier/TACO/data'
path2save = '/content/drive/MyDrive/GarbageClassifier/TACO/yolo_taco'
path2corrupt = '/content/drive/MyDrive/GarbageClassifier/TACO/yolo_taco/corrupt_annotations'
categories_indc = {'glass':0, 'metal':1, 'paper':2, 'plastic':3}
phase = 'train'

corrupt_imgs = 0
file_idx = 0

for path2img,coords in bboxes.items():
  try:
    file_idx += 1
    if file_idx > 1250:
      phase = 'val'
    img = cv2.imread(os.path.join(dir,path2img))
    img_name = Path(path2img).stem
    with open(os.path.join(path2save,phase,'labels',img_name)+'_{}.txt'.format(file_idx),'w') as f_txt:
      for points in coords:
        points = [ int(v) for v in points]
        category = categories[points[0]]['name']
        super_category = categ_map[category]
        if super_category == 'background':
          continue
        category_idx = categories_indc[super_category]
        x_t_l,y_t_l,a_width , a_height = points[1],points[2],points[3],points[4]
        a_x_centre = int(x_t_l+(a_width)/2)
        a_y_centre = int(y_t_l+(a_height)/2)
      # normalize bbox coordinates relatively to image size
        x_centre = a_x_centre/img.shape[1]
        y_centre = a_y_centre/img.shape[0]
        width = a_width/img.shape[1]
        height = a_height/img.shape[0]
        for idx,spec in enumerate([x_centre,y_centre,width,height]):
          if spec > 1:
            corrupt_imgs += 1
            print('img name:{} height:{} width:{}'.format(path2img,img.shape[0],img.shape[1]))
            print('corupt coord:{} Absolute value:{} Normalized value:{} object:{}'.format(idx,[a_x_centre,a_y_centre,a_width,a_height][idx],spec,category))
            raise Corrupt_annot
        line2write = ' '.join([str(category_idx),str(x_centre),str(y_centre),str(width),str(height)]) + '\n'
        f_txt.write(line2write)
    cv2.imwrite(os.path.join(path2save,phase,'images',img_name)+'_{}.jpg'.format(file_idx),img)
  except Corrupt_annot:
    os.remove(os.path.join(path2save,phase,'labels',img_name)+'_{}.txt'.format(file_idx))
    img = cv2.imread(os.path.join(dir,path2img))
    for bbox in coords:
        category = categories[bbox[0]]['name']
        points = [(int(bbox[1]),int(bbox[2])),(int(bbox[1]+bbox[3]),int(bbox[2])),(int(bbox[1]+bbox[3]),int(bbox[2]+bbox[4])),(int(bbox[1]),int(bbox[2]+bbox[4]))]
        cv2.polylines(img, pts = [np.array(points)], isClosed = True, color = (0, 200, 100), thickness = 4)
        img = cv2.putText(img,category ,(points[0][0],points[0][1]+30), cv2.FONT_HERSHEY_SIMPLEX,1, (255, 0, 0), 2, cv2.LINE_AA)
        cv2.imwrite(os.path.join(path2corrupt,'images',img_name)+'_{}.jpg'.format(file_idx),img)
    continue


print('overall corrupt number - {}'.format(corrupt_imgs))
```

```python
# Convert back and inspect normalized annotation values
import random
import cv2
import numpy as np

dir = '/content/drive/MyDrive/GarbageClassifier/TACO/yolo_taco'
path2save = '/content/drive/MyDrive/GarbageClassifier/TACO/yolo_taco/inspect_normalized_annots'

phase = 'train'
for i in range(20):
  paths2imgs = os.listdir(os.path.join(dir,phase,'images'))
  random.shuffle(paths2imgs)
  for img_name in paths2imgs:

    if i > 15:
      phase = 'val'
    cv_img = cv2.imread(os.path.join(dir,phase,'images',img_name))
    annot = os.path.join(dir,phase,'labels',img_name.split('.')[0]+'.txt')
    with open(annot,'r') as txt:
      lines = txt.readlines()
    for single_line in lines:
      line = [ float(v) for v in single_line.split()]

      width,height = int(line[3]*cv_img.shape[1]),int(line[4]*cv_img.shape[0])
      x_c,y_c = int(line[1]*cv_img.shape[1]),int(line[2]*cv_img.shape[0])
      t_l,t_r = (x_c-int(width/2),y_c-int(height/2)),(x_c+int(width/2),y_c-int(height/2))
      b_r,b_l = (x_c+int(width/2),y_c+int(height/2)),(x_c-int(width/2),y_c+int(height/2))
      points = np.array([t_l,t_r,b_r,b_l])

      cv2.polylines(cv_img, pts = [points], isClosed = True, color = (0, 200, 100), thickness = 7)
    cv2.imwrite(os.path.join(path2save,img_name),cv_img)
```
